```python
#Q1. Explain why we have to use the Exception class while creating a Custom Exception.
#Note: Here Exception class refers to the base class for all the exceptions.
'''
In Python, exceptions are used to handle errors and exceptional conditions that may occur during the execution of a program. The Exception class is the base class for all built-in exceptions in Python, and it provides a common interface for handling exceptions.

When creating a custom exception, we need to inherit from the Exception class (or one of its subclasses) to ensure that our custom exception can be caught and handled using the standard exception handling mechanisms in Python. This allows us to define our own exception types that can be used to signal specific error conditions or exceptional situations in our code.'''


```

```python

# Suggestion: Add a code example of creating a custom exception
class CustomException(Exception):
    """A custom exception for signaling invalid input."""

    def __init__(self, message="pwskills input."):
        """Initialize the custom exception with a message."""
        self.message = message
        super().__init__(self.message)

# Example usage:
try:
    raise CustomException()
except CustomException as e:
    print(e)
```

```python
import pprint

def print_exception_hierarchy(exc, level=0):
    """Print the exception hierarchy starting from the given exception class.

    Args:
    exc (type): The exception class to start the hierarchy from.
    level (int): The current level of indentation.
    """
    for sub_exc in exc.__subclasses__():
        print("  " * level + sub_exc.__name__)
        print_exception_hierarchy(sub_exc, level + 1)

if __name__ == "__main__":
    print("Python Exception Hierarchy:")
    print_exception_hierarchy(Exception)
```
